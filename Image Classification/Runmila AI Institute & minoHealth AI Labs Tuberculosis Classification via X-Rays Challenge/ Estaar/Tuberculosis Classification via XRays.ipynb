## Set seed for code reproducability

#### Use num_workers as 0 to maintain code reproducability when creating the loader

## Create data loaders for working on the data.

```python
# Create our dataloaders
dataloaders = ImageDataLoaders.from_df(train, label_col='LABEL', item_tfms=Resize(224, ResizeMethod.Squish), bs=8, num_workers=0)
dataloaders.show_batch()
```

## Create the learner for running a CNN architecture

```python
# random_seed(1234, True)
learn = cnn_learner(dataloaders, resnet152, metrics=[error_rate, accuracy])
```

## Train the CNN learner

```python
# random_seed(1234, True)
learn.fine_tune(25)
```

## Load the Sample Submission and test files.

Load the sample submission and modify the file to be read for prediction.

```python
sample_submission = pd.read_csv('SampleSubmission.csv')
sample_submission.head()
```

```python
test = pd.read_csv("Test.csv")
test.head()
```

## Predict the labels for the test data (Sample Submission)

```python
# Create a testing data loader
testing_dataloader = learn.dls.test_dl(test)

#generate predictions
predictions, _ = learn.tta(dl=testing_dataloader)
```

```python
# Generate labels for the sample submission
sample_submission['LABEL'] = predictions.numpy()[:, 1]
sample_submission.head()
```

## Export CSV for submission

```python
sample_submission.to_csv('sample_submission_4_6.csv', index=False)
```
