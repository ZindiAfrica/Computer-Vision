```python
#Import libraries to load and process data
import numpy as np
import pickle
import pandas as pd
from tqdm import tqdm as tqdm
```

```python
# Getting all the data
dic_types={11: "electron", 13 : "muon", 211:"pion", 321:"kaon",2212 : "proton"}
X = []
y = []

for i in tqdm(range(1,351,1)):
    fid = f'D:/TIC-HEAP Cirta/Events CERN/event{i}.pkl' #Use Directory With Event PKL files on own computer
    try:
        pkl_file = open(fid, 'rb')
        event1 = pickle.load(pkl_file)
        # get the data and target
        data,target=event1[0],event1[1]
        target=target.astype('int')
        X += [d for d in data]
        y += [t for t in target]
    except:
        print('problem with file', i)
```

```python
df = pd.DataFrame({'y':y})
df['class'] = df['y'].map(dic_types)
for i in range(100):
    df[str(i)] = [x.flatten()[i] for x in X]
df.sample(5)
```

```python
df.to_csv('train.csv')
```
