```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
sub1 = pd.read_csv('se_resnext101_32x4d.csv')
sub2 = pd.read_csv('senet154_100.csv')
sub3 = pd.read_csv('senet154_50.csv')
```

```python
sub1.head()
```

```python
ids = sub1["image"]
columns = sub1.columns
```

```python
sub1 = sub1.iloc[:, 1:]
sub2 = sub2.iloc[:, 1:]
sub3 = sub3.iloc[:, 1:]
```

```python
trim_mean = stats.trim_mean([sub1.values, sub2.values, sub3.values], 0.2)
trim_mean
```

```python
sub = pd.DataFrame(np.stack([ids, trim_mean[:,0]], axis=1), columns=columns)
```

```python
sub.to_csv("avg.csv", index=False)
```
