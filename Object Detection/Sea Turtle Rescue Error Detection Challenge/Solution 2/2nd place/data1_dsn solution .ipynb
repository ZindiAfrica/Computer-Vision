I find values in Fisher column of dirty dataframe that are not in Fisher column in clean dataframe and I match those
values to values in Fisher column in test dataframe. The idea is any value found in Fisher column of dirty dataframe and not found in Fisher column of clean dataframe is considered erroneous, and if these erroneous values are still found in the test dataframe then these values are erroneous.    

The below codes carry out the process.

```python
#search for values in dirty dataframe and not in clean dataframe
fisherdc = dirty["Fisher"][~dirty["Fisher"].isin(clean["Fisher"])]
fisherdc.unique()
```

```python
#search for values in test dataframe and not in clean dataframe
db  = test["Fisher"][~test["Fisher"].isin(clean["Fisher"])]
db.unique()
```

```python
#values in dirty dataframe and in test dataframe which are considered erroneous
user = db[db.isin(fisherdc)]
user.unique()
```

```python
#check position of the values in the test dataframe
array = ['fisher_581', 'fisher_1385', 'fisher_1024', 'fisher_815',
       'fisher_850', 'fisher_583', 'fisher_1318', 'fisher_579',
       'fisher_985', 'fisher_1319']
dx = test.loc[test["Fisher"].isin(array)]
dx[['Rescue_ID', 'Fisher']]
```

End    
fisher values above are detected error

create a dataframe to store Fisher values detected as error

```python
#add string to Rescue_ID values
dx['Rescue_ID'] = dx['Rescue_ID'].astype(str) + ' ' + 'x' + ' ' + 'Fisher'
```

```python
#create a column 'error' with 1 all through to indicate error
dx['error'] = 1
```

```python
dx = dx[['Rescue_ID', 'Fisher', 'error']]
```

```python
#csv file
dx.to_csv('fishervalueswitherror.csv', index=False)
```

```python
#End of creating dataframe to store Fisher values detected as error
```
